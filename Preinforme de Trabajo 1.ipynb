Consulte cuáles son los pilares de la programación orientada a objetos, defínalos en sus palabras y realice un ejemplo práctico para cada uno, comentando en qué parte se aplican. Los pilares básicos son: Abstracción, Herencia, Polimorfismo y Encapsulamiento, aunque en algunos sitios reportan más.

**Solución:**
Abstracción: su principal objetivo es ocultar al usuario la complejidad del codigo. Para esto usamos metodos y atributos que se encuentran adentro de las clases, siendo estos los que ocultan la complejidad del codigo.En otras palabras, abstraemos la funcionalidad del codigo en clases, las cuales se pueden usar facilmente sin necesidad de comprender su funcionamiento.

Herencia: se utliza basicamente para reutilizar codigo, permitiendonos crear diferentes subclases de una clase previamente definida, herendando asi sus atributos y métodos.

Polimorfismo: Nos permite modificar brevemente los metodos y atributos de las subclases(los cuales ya estuvieron previamente definidos en una superclase). Basicamente podemos adaptar una subclase a nuestras necesidades.

Encapsulamiento: se utiliza para proteger los datos de una clase, y de esta manera poder regular el acceso a la informacion por medio de metodos que proporcina la clase(setters y getters)
